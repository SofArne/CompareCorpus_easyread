## Her prøver vi oss med Type Token Ratio

Type token ratio er en standard måte å måle leksikalsk variasjon. Forholdet beregnes ved å dividere antall typer ord på antallet forekomster: $ttr\leftarrow\frac{antall\ typer}{antall\ token\ (forekomster)}$. For en stor tekst vil ttr være lavere enn for en mindre tekst, fordi på et eller annet tidspunkt går man tom for nye ord, så de gamle brukes på nytt, samt at funksjonsord blir brukt om og om igjen, noe som vil trekke ttr ned. For å kunne sammenligne tekster, korte mot lange, deles tekstene inn i bolker (*chunks*) og ttr er snittet av ttr for hver bolk. 

### Import av ressurser

```python
import nbtext as nb
from nbtext import get_urns_from_text, Corpus_urn, get_urn, sttr

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def histogram(f, r, labels=['lettlest','referanse'], title="Sammenlign"):
    plt.figure(figsize=(10,5))
    plt.title(title)
    lett = plt.hist(f, bins=10, density=True, color='blue', alpha=0.6, label ='lett');
    ref = plt.hist(r, bins=10, density=True, color='orange', alpha = 0.6, label ='referanse');
    plt.legend(labels)
    plt.tight_layout()
    return
```

## URN-ene fra lettlestbøker og referanse hentes inn

```python
# finn rett sti til mappen med word-filer, og erstatt den med den som ligger her

urns = get_urns_from_text('./urner_lettlest.txt')

# lager et referansekorpus

refurns = get_urns_from_text('./urnref.txt')
refurns = {r for r in refurns if not r.startswith('2017') and not r.startswith('2018')}
```

```python
# lager et referansekorpus

refurns_historie = nb.get_urn({'year':1970, 'next':60, 'ddk':"901%", 'limit':200})
reference = [r[0] for r in refurns_historie]
```

# Beregner STTR

Dataene brukes i sammenligningen kommer fra lister over URNer som ligger i fila ´urner_lettlest.txt´ og ´urnref.txt´, i tillegg til Dewey.

WebDewey ligger [her](http://deweysearchno.pansoft.de/webdeweysearch/executeSearch.html?lastScheduleRecord=&lastTableRecord=&query=808.80384&catalogs=nb.no)

```python
u_ttr = {u:sttr(u) for u in urns} # lettlestbøker
r_ttr = {u:sttr(u) for u in refurns} # ungdomsbøker
rf_ttr = {u:sttr(u) for u in reference} # fra dewey
```

# STTR

Her tas ut 5000 ord i gangen - 5000 ord er bolkstørrelsen ved default.

Fra diagrammene ser det ut som at det ikke er så stor forskjell mellom ungdomsbøkene og lettlestbøkene i ttr-målet for kompleksitet. Tvert om, kan det virke som lettlestbøkene er mer kompleks. Men variasjonen er ganske stor.

```python
histogram(u_ttr.values(), r_ttr.values(), labels=['lettlest','ungdomsbøker'], title='Lettlest og ungdomsøker')
```

```python
histogram(u_ttr.values(), rf_ttr.values(), labels=['lettlest','dewey 901'], title='Lettlest og dewey 901')
```

# Prøver en annen chunking

Lavere chunking gir høyere TTR. En sammenligning av snittverdiene diagrammet under med de over, ser vi at ttr har økt, men forskjellene er de samme mellom sammenligningstekstene. Lettlestbøkene viser ingen større kompleksitet, mens dewey-utvalget gjør det.

```python
u_ttr_2 = {u:nb.sttr(u, chunk=2000) for u in urns}
r_ttr_2 = {u:sttr(u, chunk=2000) for u in refurns}
rf_ttr_2 = {u:sttr(u, chunk= 2000) for u in reference}
```

```python
histogram(u_ttr_2.values(), r_ttr_2.values(), labels=["lettlest", "ungdomsbøker"], title="Ordbolk 2000 - lettlest mot ungdom")
```

```python
histogram(u_ttr_2.values(), rf_ttr_2.values(), labels=["lettlest", "dewey"], title="Ordbolk 2000 - lettlest mot dewey")
```
